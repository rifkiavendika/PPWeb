```python
import pandas as pd
```

```python
politik = pd.read_csv('/content/drive/MyDrive/PPWeb/tugas/berita_detik.csv')
# Menambahkan kolom 'kategori' dengan nilai 'Politik'
politik = politik.assign(Kategori=['Politik'] * len(politik))
politik.head()
```

```python
olahraga = pd.read_csv('/content/drive/MyDrive/PPWeb/tugas/berita_detik2.csv')
# Menambahkan kolom 'kategori' dengan nilai 'olahraga'
olahraga = olahraga.assign(Kategori=['Olahraga'] * len(olahraga))
olahraga.head()
```

```python
kesehatan = pd.read_csv('/content/drive/MyDrive/PPWeb/tugas/berita_detik3.csv')
# Menambahkan kolom 'kategori' dengan nilai 'kesehatan'
kesehatan = kesehatan.assign(Kategori=['Kesehatan'] * len(kesehatan))
kesehatan.head()
```

```python
# Menggabungkan ketiga DataFrame menjadi satu DataFrame
combined_df = pd.concat([politik, olahraga, kesehatan], axis=0)

# Mengacak baris-baris dari DataFrame yang telah digabungkan
all_kategori = combined_df.sample(frac=1).reset_index(drop=True)
all_kategori = all_kategori.drop('Unnamed: 0', axis=1)
```

```python
# Simpan DataFrame yang telah digabung dan diacak ke dalam file CSV
all_kategori.to_csv('/content/drive/MyDrive/PPWeb/tugas/all_kategori.csv', index=False)
```
